### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
data = datasets.load_iris()
```

```python
X = data.data
```

```python
y = data.target
```

```python
print(X[:5, :], "\n\n")
print(y[:5])
```

```python
#Transform 4 dimensioned data  to 3 dimensions using principal component analysis

pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)
```

```python
print(X[:5, :], "\n\n")
```

```python
fig = plt.figure(figsize = (8,6))

ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean() + 1.8,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=0.1, edgecolor='w', facecolor='w'))

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.rainbow, edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
